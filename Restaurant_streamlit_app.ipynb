```python
%%writefile app.py
import streamlit as st
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.impute import SimpleImputer

cleaned_data = pd.read_csv("cleaned_data.csv")
encoded_data = pd.read_csv("encoded_data.csv", on_bad_lines='skip')

# Select numeric columns
numeric_columns = encoded_data.select_dtypes(include=['float64', 'int64']).columns

# Impute missing values if any
imputer = SimpleImputer(strategy='mean')
encoded_data[numeric_columns] = imputer.fit_transform(encoded_data[numeric_columns])


# Perform K-Means clustering on the numeric data
kmeans = KMeans(n_clusters=5, random_state=42)
encoded_data['cluster'] = kmeans.fit_predict(encoded_data[numeric_columns])


# Streamlit App Title
st.title("🍽️ Restaurant Recommendation System")

#NAVIGATING TO THE TABS
navigation = st.sidebar.radio('Navigate to:',["Home", "User Preferences", "Restaurant Picks"])

if navigation == "Home":

    st.header("Welcome to the Restaurant Recommendation System!")
    st.write(""" Are you craving something delicious but not sure where to go? You've come to the right place!
    Our Restaurant Recommendation System is designed to help you discover the perfect dining spot tailored to your preferences.
    Whether you're in the mood for a cozy café, a fine dining experience, or a quick bite, we’ve got you covered.
    """)

# Cleaned Dataset Page
elif navigation == "User Preferences":
    st.header("Enter Your Preferences")
    city = st.selectbox("City", cleaned_data['city'].unique())
    cuisine = st.selectbox("Cuisine", cleaned_data['cuisine'].unique())
    rating = st.slider("Minimum Rating", 1.0, 5.0, 4.0)
    cost = st.selectbox("Cost Range", cleaned_data['cost'].unique())

    if st.button("Get Restaurants"):
    # Filter restaurants based on user preferences
      filtered_restaurants = cleaned_data[
          (cleaned_data['city'] == city) &
          (cleaned_data['cuisine'] == cuisine) &
          (cleaned_data['cost'] == cost) &
          (cleaned_data['rating'] >= rating)
      ]

      if not filtered_restaurants.empty:
          # Get the indices of the filtered restaurants in cleaned_data
          filtered_indices = filtered_restaurants.index
          # Use these indices to get the corresponding clusters from encoded_data
          filtered_clusters = encoded_data.iloc[filtered_indices]['cluster'].unique()

          # Recommend restaurants in the same clusters
          recommended_restaurants = cleaned_data[cleaned_data['name'].isin(encoded_data[encoded_data['cluster'].isin(filtered_clusters)]['name'])]
          recommended_restaurants = recommended_restaurants[recommended_restaurants['rating'] >= rating]


          # Display recommended restaurants
          st.header("Recommended Restaurants")
          st.dataframe(recommended_restaurants[['name', 'rating', 'rating_count', 'address']])
      else:
          st.write("No restaurants match your preferences.")


elif navigation == "Restaurant Picks":
    st.header("Recommend Restaurants Based on Dropdown Selection")
    restaurant_name = st.selectbox("Select a Restaurant", cleaned_data['name'].unique())

    if restaurant_name:
        # Find the cluster for the selected restaurant
        selected_restaurant = cleaned_data[cleaned_data['name'] == restaurant_name]

        # Display recommended restaurants
        # Display the selected restaurant details
        st.header("Restaurant Picks")
        st.dataframe(selected_restaurant[['name', 'rating', 'rating_count', 'address']])
    else:
        st.write("Please select a restaurant from the dropdown.")
```
