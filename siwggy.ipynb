```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("swiggy.csv")
data
```

```python
data.isnull().sum()
```

```python
data.drop(columns=['id', 'lic_no', 'link', 'menu'], axis=1, inplace=True)
data.columns
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
data['rating_count'] = data['rating_count'].replace('Too Few Ratings', '5')

def convert_rating_count(value):

    value = value.replace(' ratings', '').strip()
    if '+' in value:
        value = value.replace('+', '')  # Remove '+' sign
    if 'K' in value:
        value = float(value.replace('K', '')) * 1000  # Convert 'K' to thousands
    return int(value)

# Apply the conversion function
data['rating_count'] = data['rating_count'].apply(convert_rating_count)
```

```python
data['rating_count'].isnull().sum()
```

```python
data['rating'].value_counts()
```

```python
data['rating'] = data['rating'].replace('--', '0.0')
```

```python
data['rating'] = pd.to_numeric(data['rating'], errors='coerce')
data['rating'] = data['rating'].astype(float)
```

```python
data['rating'].isnull().sum()
```

```python
data['cost'] = data['cost'].str.replace(r'[₹\s]', '', regex=True)  # Remove currency symbol and spaces
data['cost'] = pd.to_numeric(data['cost'], errors='coerce').astype(int)
```

```python
data['cost'].isnull().sum()
```

```python
data.dtypes
```

```python
data.head()
```

```python
column_names = ['name', 'city', 'rating','rating_count', 'cost', 'cuisine', 'address']
data.to_csv('cleaned_data.csv', columns=column_names, index=False)
```

```python
cleaned_data = pd.read_csv('cleaned_data.csv')
cleaned_data.head()
```

# Encoding

```python
from sklearn.preprocessing import OneHotEncoder
import pickle

categorical_columns = ['name', 'city', 'cuisine']

# Create OneHotEncoder object for remaining categorical features
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Fit and transform the categorical columns
encoded_data = encoder.fit_transform(cleaned_data[categorical_columns])
```

```python
# Convert the encoded data into a DataFrame
encoded_columns = encoder.get_feature_names_out(categorical_columns)
encoded_df = pd.DataFrame(encoded_data, columns=encoded_columns)
```

```python
# Step 3: Save the encoder as a Pickle file
with open('encoder.pkl', 'wb') as f:
    pickle.dump(encoder, f)
```

```python
final_data = pd.concat([cleaned_data, encoded_df], axis=1)
```

```python
# Step 5: Save the preprocessed dataset as encoded_data.csv
final_data.to_csv('encoded_data.csv', index=False)
```

```python
print("Indices of cleaned_data and encoded_data match:", final_data.index.equals(cleaned_data.index))
```
